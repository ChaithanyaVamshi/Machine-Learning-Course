#Definition of some professional terms:
- Null hypothesis (H0)

The null hypothesis is the initial assumption considered to be true.

- Alternative hypothesis (H1) 

An alternative hypothesis is simply opposite of null hypothesis. If the result from the calculation is inferred that null hypothesis shall be rejected, the alternative hypothesis would be supported instead.

- P_value 

P-value is the probability for a given statistical model served as a checkpoint whether to accept or reject the null hypothesis. If the p-value is less than or equal to the level of significance, the null hypothesis shall be overthrown.
- Significance Level:

Itâ€™s a likelihood of observing how strong the evidence is. When the p-value is higher than the threshold, the null hypothesis would not be rejected. Otherwise, the alternative hypothesis would be rejected when the p-value is lower than the threshold.

    We choose to calculate a Pearson correlation coefficient and the p-value for testing correlation from scipy package.
    The Pearson correlation coefficient measures the linear relationship between two datasets with the value ranged form -1 and 1.
    The value of -1 indicates the negative linear relationship while +1 is implied with positive linear relationship and 0 is shown with no correlation.

The code of the Pearson correlation coefficient measure is pasted below ! First I transferred the dataset into a Pandas DataFrame. Convert Pandas Dataframe column into arrays and pass them to Pearson scipy package for p_value.

```
import pandas as pd
from scipy import stats

df = pd.read_csv('house_price.csv')
X = df['YearBuilt'].array
Y = df['SalePrice'].array
result = stats.pearsonr(X,Y)
print(result)
## (Pearsonâ€™s correlation coefficient, Two-tailed p-value)
```

#Decide whether to reject or fail to reject the null hypothesis and make the decision

As the p-value is shown from the Pearson Correlation test, the p-value is way less than significance level of 0.05. Therefore, we shall reject null hypothesis based on the p-value output from Pearson test.


From Pearson 2-tailed value, when p-value falls into the area out of the green zone, the null hypothesis shall be rejected. On the other hand, when the p-value falls into green area, the null hypothesis would be accepted.
